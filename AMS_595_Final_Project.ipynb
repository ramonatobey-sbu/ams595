```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('MVC.csv')
#display(df)
```

```python
filtered_data = pd.read_csv('MVC.csv', usecols=['CRASH DATE', 'LATITUDE', 'LONGITUDE', 'ON STREET NAME', 'CROSS STREET NAME', 'NUMBER OF PERSONS INJURED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2'])
```

```python
# Graphing Crash Locations by Latitude and Longitude

# remove all latitude and longitude
filtered_data_cleaned = filtered_data[(filtered_data['LATITUDE'] != 0) | (filtered_data['LONGITUDE'] != 0)]


plt.figure(figsize=(10, 8))
sns.scatterplot(x='LONGITUDE', y='LATITUDE', data=filtered_data_cleaned, alpha=0.5, hue='NUMBER OF PERSONS INJURED', palette='dark')
plt.title('Crash Locations by Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
# take out columns where the street names are empty
filtered_data_cross_streets = filtered_data[(filtered_data['ON STREET NAME'] != '') | (filtered_data['CROSS STREET NAME'] != '')]

# Clean street names
filtered_data['ON STREET NAME'] = filtered_data['ON STREET NAME'].astype(str).str.strip().str.upper()
filtered_data['CROSS STREET NAME'] = filtered_data['CROSS STREET NAME'].astype(str).str.strip().str.upper()

# add a column for intersections to the filtered data
filtered_data['INTERSECTION'] = filtered_data.apply(lambda row: ' & '.join(sorted([row['ON STREET NAME'], row['CROSS STREET NAME']])) if row['CROSS STREET NAME'] and row['CROSS STREET NAME'] != 'NAN' else row['ON STREET NAME'], axis=1)

# count the number of crashes at each intersection pair
intersection_counts = filtered_data['INTERSECTION'].value_counts()

# print out the top 20 intersections for number of crashes
top_20_intersections = intersection_counts.head(20)
print("Top 20 Intersections with Most Crashes:")
print(top_20_intersections)
```

```python
# Graph top 5 intersections and their crash type in different colors

# FIRST INTERSECTION: MAJOR DEEGAN EXPRESSWAY & WEST FORDHAM ROAD
# isolating the data for the first intersection
intersection1 = filtered_data[filtered_data['INTERSECTION'] == 'MAJOR DEEGAN EXPRESSWAY & WEST FORDHAM ROAD']
intersection1_factor_counts = intersection1['CONTRIBUTING FACTOR VEHICLE 1'].value_counts()

# make a bar graph out of intersection1_factor_counts
plt.figure(figsize=(12, 8))
sns.barplot(x=intersection1_factor_counts.index, y=intersection1_factor_counts.values)
plt.title(f'Contributing Factors for Crashes at {'MAJOR DEEGAN EXPRESSWAY & WEST FORDHAM ROAD'}')
plt.xlabel('Contributing Factor Vehicle 1')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()

```
